```python
import pandas as pd
```

```python
df = pd.read_excel('./data/0_科系_學院分類.xlsx')
```

```python
df.head(2)
```

```python
df_ex = pd.read_excel('./data/0_科系_學院分類_例外.xlsx').dropna(subset=['科系'])
```

```python
df_ex.head(2)
```

```python
for col in df.columns:
    if col not in ['院','系所']:
        print(col)
        confirm_dict = dict(zip(df['系所'],df[col]))
        df_ex[col] = df_ex['科系'].apply(lambda x:'待定' if x not in confirm_dict.keys() else confirm_dict[x])
```

```python
df_ex.to_excel('./data/0_科系_學院分類_例外.xlsx', index=False,encoding='utf-8-sig')
```
