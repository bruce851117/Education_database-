```python
df = pd.read_excel('C:\\Users\\Nancy\\NTU\\graduate\\研究室\\論文\\coding\\ntuFinance\\coding\\data\\2_高階經理人_M2800 金融業_0423.xlsx')
df.head()
```

```python
dfCount = df[['公司代碼簡稱','資料源年','總經理_燈號',
                  '財務長_燈號','營運長_燈號','法遵長_燈號','法務長_燈號','技術長_燈號',
                  '資訊長_燈號','資安長_燈號']]
dfCount
```

```python
df_all = pd.DataFrame()
for position in ['總經理','財務長','營運長','法遵長','法務長','技術長','資訊長','資安長']:
    df2 = pd.pivot_table(dfCount,
                   index=[position + '_燈號'],
                   values=['公司代碼簡稱'],
                   aggfunc=lambda x: len(x)
                  )
    df2.columns = [position]
    df_all = pd.concat([df_all,df2],axis = 1)
```
